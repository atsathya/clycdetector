```python
def gimme_res(data, index_lower):
  max_value = max(data['0'][index_lower:])
  max_index = np.argmax(data['0'][index_lower:])+index_lower

  half_max_value = max_value / 2

  left_index = np.argmin(np.abs(data[index_lower:max_index] - half_max_value))+index_lower
  right_index = np.argmin(np.abs(data[max_index:] - half_max_value)) + max_index

  fwhm = right_index - left_index
  return(round(((fwhm/(max_index))*100),2))
```

```python
res_Na_1 = gimme_res(df_Na,2500)
res_Na_2 = gimme_res(df_Na,6800)
res_Cs = gimme_res(df_Cs,3000)
res_Ba = gimme_res(df_Ba,1800)
res_Co_2 = gimme_res(df_Co,7400)
res_Co_1 = gimme_res(df_Co,6500)
```

```python
x_2 = [ 356,511, 661, 1173.2, 1332]
y_2 = [ res_Ba,res_Na_1, res_Cs, res_Co_1, res_Co_2]
```

```python
from scipy.interpolate import interp1d, pchip
x_smooth = np.linspace(x_2[0], x_2[-1], 500)
f1 = interp1d(x_2, y_2, kind='quadratic')
fig, ax = plt.subplots()
ax.plot(x_smooth, f1(x_smooth))

fig.set_size_inches(10, 5)
plt.xlabel("√(Energy(KeV))", fontsize=16)
plt.ylabel("Resolution(%)", fontsize=16)
ax.legend(['Actual'],loc='best', fontsize='large')
plt.title("CLYC - Energy Vs Resolution", fontsize=20, weight="bold")
plt.show()
```
